METHOD

    loc[ ]. Éste permite "localizar" elementos en el DataFrame por índices (o condiciones) en las filas, y por nombres de columnas (o condiciones) en las columnas.

    iloc[ ]. Éste permite "localizar" elementos por índices de fila y columna


```python

df.loc[10:20, ['ID', 'income']]
```

```python
df.iloc[10:20, [0, 2]]
```

También podemos utilizar el método .loc para filtrar de la forma que vimos antes y seleccionar únicamente aquellas columnas que nos interesen

```python
df.loc[df['education_level'] == 2, ['ID', 'age', 'income']]
```
